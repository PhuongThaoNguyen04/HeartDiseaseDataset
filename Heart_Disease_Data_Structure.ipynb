# **Missing Values**
In this analysis, we identified the number of missing values across the dataset using *df.isnull().sum()*. We calculated the percentage of missing values for each features to assess their impact. For considered:


*   **Removing rows with missing data**, which is effective when the percentage of missing values is small.
*   **Imputation**, where the mean value of a numerical column was used to fill in the missing data for *age* and *resting_blood_pressure*. This method preserves the dataset size and is suitable when the missing data is random and not significant in quantity.

We found that 0.99%, 1.65% of the dataset had missing values in the *age* and *resting_blood_pressure* columns. Imputing these values allowed us to retain a larger dataset for analysis while minimimizing data loss.

# **Identify Outliers**

Outliers were detected using the **Interquartile Range** method, which identifies outliers as values that fall below or above. Additionally, we applied the **Z-score** method to find data points more than three standard deviations from the mean.

The columns *resting_blood_pressure*, *cholesterol*,*fasting_blood_pressure*, *maximum_heart_rate*, *oldpeak*, *ca* exhibited significant outliers. These outliers were either removed or capped to fall within acceptable ranges, depending on their potential influence on the analysis. In practice, some outliers may contain important insights, and careful consideration is needed before removing them.











```python
import numpy as np
from scipy import stats

# Applying Z-score method to detect outliers
# Z-scores measure how many standard deviations a data point is from the mean.
# Applying Z-score > 3 or < -3 is considered an outlier.
z_scores = np.abs(stats.zscore(df.select_dtypes(include=['float64', 'int64'])))
outliers_z = (z_scores > 3).sum(axis=0) # COunt of outliers per column
outliers_z

# Z-score is suitable here because it effectively identifies extreme values
# in approximately normally distributed data, such as age and blood pressure.
```

```python
# IQR method
# Outliers are data points that lie outside 1.5*IQR from the quartiles.
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Outliers are points below Q1 - 1.5*IQR or above Q3 + 1.5*IQR
# Identifying outliers (True for outliers, False otherwise)
outliers = (df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))
outliers_count = outliers.sum()
print("Outliers Count per Column:")
print(outliers_count)
# IQR method is useful for skewed distributions but may miss extreme outliers in normal data.
```

```python
# Handling Missing Values
# Option 1: Remove rows with missing values
df_removed = df.dropna()

# Option 2: Impute missing values using the mean for numerical columns
df_imputed = df.copy()
df_imputed['age'].fillna(df_imputed['age'].mean(), inplace=True)
df_imputed['resting_blood_pressure'].fillna(df_imputed['resting_blood_pressure'].mean(), inplace=True)

# Checking the dataset shape after removing or imputing missing values
print("Shape after removing rows with missing values:", df_removed.shape)
print("Shape after imputing missing values:", df_imputed.shape)

# Missing values are handled by removal or imputation, high cardinality is identified,
# and outliers are detected using both IQR (for skewed data) and Z-score (for normal distributions).
```
